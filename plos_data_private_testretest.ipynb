```python
import pandas as pd
import pylab as pl
%matplotlib inline
```

```python
mdft = pd.read_csv('test_retest_mdf_df_romm.csv')
```

```python
pl.plot([1,2,3,4],mdft[['F1_T1','F1_T2','F1_T3','F1_T4', 'F1_T5']].corr()[0:1].values[0][1:],'-o',label='F1')
pl.plot([1,2,3,4],mdft[['F2_T1','F2_T2','F2_T3','F2_T4', 'F2_T5']].corr()[0:1].values[0][1:],'-o',label='F2')
pl.plot([1,2,3,4],mdft[['F3_T1','F3_T2','F3_T3','F3_T4', 'F3_T5']].corr()[0:1].values[0][1:],'-o',label='F3')
pl.plot([1,2,3,4],mdft[['F4_T1','F4_T2','F4_T3','F4_T4', 'F4_T5']].corr()[0:1].values[0][1:],'-o',label='F4')
pl.plot([1,2,3,4],mdft[['F5_T1','F5_T2','F5_T3','F5_T4', 'F5_T5']].corr()[0:1].values[0][1:],'-o',label='F5')
pl.legend(loc='best')
ax=pl.axes()
ax.set_xticklabels([1,' ', 2,' ' , 3,' ' , 4])
# Shrink current axis's height by 10% on the bottom
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.tick_params(axis='x',which='minor',bottom='off')
# Put a legend below current axis
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),fancybox=True, shadow=True, ncol=5)
pl.xlabel('Time period', fontsize=14)
pl.ylabel('Correlation', fontsize=14)
```
